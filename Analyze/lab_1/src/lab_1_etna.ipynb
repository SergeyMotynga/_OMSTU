```python
import matplotlib.pyplot as plt
import pandas as pd
from etna.datasets import TSDataset
```

```python
gold_df = pd.read_excel("gold.xlsx")
gold_df.columns = ["timestamp", "target"]
gold_df["segment"] = "gold"

silver_df = pd.read_excel("silver.xlsx")
silver_df.columns = ["timestamp", "target"]
silver_df["segment"] = "silver"

both_df = pd.concat([gold_df, silver_df], ignore_index=True)

gold_df.head()
silver_df.head()
both_df.head()
```

```python
gold_ts = TSDataset(df=TSDataset.to_dataset(gold_df), freq="D")
silver_ts = TSDataset(df=TSDataset.to_dataset(silver_df), freq="D")
both_ts = TSDataset(df=TSDataset.to_dataset(both_df), freq="D")
```

```python
gold_ts.info()
gold_ts.describe()
gold_ts.plot()
```

```python
silver_ts.info()
silver_ts.describe()
silver_ts.plot()
```

```python
both_ts.info()
both_ts.describe()
both_ts.plot()
```
